```python
import pandas as pd
```

```python
df=pd.read_csv("games.csv")
df
```

```python
df.dropna(how='all', inplace=True)
df
```

```python
# df.drop(columns=["img" , "critic_score2"])
df = df.loc[:, ~df.columns.str.startswith(tuple(['Unnamed',"img","critic_score2"]))]
df
```

```python
df = df.copy()

df.drop_duplicates(inplace=True)
df
```

```python

df["developer"]=df["developer"].str.strip("123//")
df
```

```python

df[["consol", "Genre"]] = df["genre"].str.split("/", expand=True)
df=df.drop(columns="genre")
df
```

```python
df[["critic_score", "jp_sales"]] = df[["critic_score", "jp_sales"]].fillna("")
df
```

```python
df=df.copy()
print(df.dtypes)

```

```python
df["critic_score"] = pd.to_numeric(df["critic_score"], errors="coerce")
df["jp_sales"] = pd.to_numeric(df["jp_sales"], errors="coerce")
print(df.dtypes)

```

```python


df["critic_score"] = df["critic_score"].fillna(df["critic_score"].mean())
df["jp_sales"]=df["jp_sales"].fillna(df["jp_sales"].mean())
df
```

```python

df.rename(columns={
    "na_sales" :"north_america_sales",
    "jp_sales" :"japan_sales",
    "pal_sales" : "europe_sales"
}, inplace= True)
```

```python

df = df[["title", "publisher","developer","Genre", "consol", "total_sales", "critic_score", "north_america_sales", "japan_sales", "europe_sales", "other_sales"]]
df
```
